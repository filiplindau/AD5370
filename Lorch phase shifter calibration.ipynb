## Calibration of a Lorch VP-360-3000-300-S voltage controlled phase shifter. 
Serial number: AN5

A Rohde Schwarz network analyzer was used to measure the S21 transmission data at 2998.5 MHz. The connecting cables were calibrated away so that phase and attenuation was 0 dB / 0 deg without the phase shifter connected.

#### Basic imports

```python
%matplotlib inline
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Range1d, LinearAxis
```

#### Raw data samples:

```python
U = array([0, 0.25, 0.5, 0.75, 1., 1.25, 1.5, 1.75, 2., 2.25, 2.5, 
           2.75, 3., 3.25, 3.5, 3.75, 4., 4.25, 4.5, 4.75, 5., 5.25, 
           5.5, 5.75, 6., 6.25, 6.5, 6.75, 7., 7.25, 7.5, 7.75, 8., 
           8.25, 8.5, 8.75, 9., 9.25, 9.5, 9.75, 10.])
```

```python
ph = array([174.9, -177.6, -169.4, -161.6, -152.6, -143.6, -133.1, -121.7, -108.5, 
            -93.6, -76.1, -56.0, -32.5, -5.8, 23.7, 54.8, 86.1, 115.8, 143.7, 168.9, 
            -168.2, -148.1, -130.9, -115.6, -102.7, -91.7, -82.4, -74.5, -67.7, 
            -61.9, -56.8, -52.6, -48.9, -45.2, -43.0, -40.6, -38.5, -36.7, -35.1, 
            -33.7, -32.4])
```

```python
att = array([2.4, 2.5, 2.6, 2.7, 2.7, 2.8, 3.0, 3.1, 3.2, 3.4, 3.6, 3.8, 4.0, 
             4.2, 4.4, 4.5, 4.6, 4.4, 4.2, 3.7, 3.3, 3.0, 2.7, 2.5, 2.3, 2.2, 
             2.1, 2.0, 2.0, 1.9, 1.9, 1.8, 1.8, 1.7, 1.7, 1.7, 1.7, 1.6, 1.6, 
             1.6, 1.6])
```

#### Process phase data to remove 360 deg jumps:

```python
ph0 = ph[0]
dph = ph - ph[0]
```

```python
add_ph = [0.0]
for i in range(len(dph)-1):
    new_ph = add_ph[-1]
    if abs(dph[i+1] - dph[i]) > 180.0:
        new_ph += 360.0
    add_ph.append(new_ph)
add_ph = array(add_ph)
c_ph = add_ph + dph
```

#### Plotting:

```python
output_notebook()
```

```python
p = figure(title="Phase shifter attenuation vs applied voltage", width=550)
p.line(U, c_ph, line_width=2, legend="phase")
p.extra_y_ranges = {"att": Range1d(0, 7)}
p.line(U, att, line_width=2, y_range_name="att", color="orange", 
       legend="attenuation")
p.xaxis.axis_label = "U / V"
p.yaxis.axis_label = "Phase / deg"
ax = LinearAxis(y_range_name="att", axis_label="Attenuation / dB")
p.add_layout(ax, "right")
p.legend.location = "top_left"
```

```python
show(p)
```
